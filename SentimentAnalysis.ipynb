```python
df = pd.read_csv('cleaned_data20220418.csv', encoding = 'utf-8')
df.head(3)
```

## Converting Text column to Lowercase

```python
df['full_text'] = df['full_text'].apply(lambda x: " ".join(x.lower() for x in x.split()))
df.head(3)
```

# Removing stopwords

```python
stop = stopwords.words('english')
df['full_text'] = df['full_text'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
df.head(3)
```

# Stemming

```python
st = PorterStemmer()
```

```python
df['full_text'] = df['full_text'].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))
```

```python
df.head(3)
```

## Sentiment Analysis

## Calculating Sentiment Score

First score is sentiment polarity score and second score is subjectivity score

```python
df.head(5)
```

```python
def polar(x):
    return TextBlob(x).sentiment.polarity 
def subj(x):
    return TextBlob(x).sentiment.subjectivity
    
```

```python
df['polarity'] = df['full_text'].apply(polar)
```

```python
df['subjectivity'] = df['full_text'].apply(subj)
```

```python
df.head(5)
```

```python
df = df.drop('Unnamed: 0', 1)
df.head(5)
```

```python
def senti(x):
    if 0.5<= x <= 1:
        return 'Extremely Positive'
    elif 0.25<= x < 0.5:
        return 'Positive'
    elif 0.01<= x < 0.25:
        return 'Mostly Positive'
    elif x == 0:
        return 'Neutral'
    elif -0.25<= x < 0:
        return 'Mostly Negative'
    elif -0.5<= x < -0.25:
        return 'Negative'
    elif -1<= x <-0.5:
        return 'Extremely Negative'
        
    
    
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()
def vader(x):
    return analyzer.polarity_scores(x)
def vader_comp(x):
    return analyzer.polarity_scores(x)['compound']

df['Vader_Scores'] = df['full_text'].apply(vader) 
df['Vader_Compound_Score'] = df['full_text'].apply(vader_comp)
df.head(3)
```

```python
df['Tweet_result'] = df['polarity'].apply(senti)
```
